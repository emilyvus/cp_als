## T-test on Vietnamese (KHV) and Japanese (JPT) populations

```python
g_khv_df = khv_df.groupby(["gene"]).sum().sum(axis=1)
g_khv_df = g_khv_df /len(khv_genome_ids)
khv = pd.DataFrame()
khv['mean']= g_khv_df / len(khv_genome_ids)
khv.head()
```

```python
g_jpt_df = jpt_df.groupby(["gene"]).sum().sum(axis=1)
g_jpt_df = g_jpt_df / len(jpt_genome_ids)
jpt = pd.DataFrame()
jpt['mean']= g_jpt_df / len(jpt_genome_ids)
jpt.head()
```

```python
t_statistic, p_value = stats.ttest_ind(khv['mean'], jpt['mean'], equal_var=True)
print(f"\nTwo-Sample Independent T-Test:")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```
