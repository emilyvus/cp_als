# Step 1: Split the INFO field into a dictionary

```python
series = df['INFO'].str.split(';')
```

```python
series
```

```python
type(series)
```

```python
ll = list()
for i, elem in enumerate (series):
    d=dict()
    for item in elem:
       if "=" in item:
           k,v = item.split("=")
       else:
           k,v = item, None
           print (i, elem, "Assign None")

       d[f"INFO:{k}"]= v
    # print(d)
    ll.append(d)
    #print(ll)
    # break
```

```python
len (ll)
```

```python
ll[0]
```

# Step 2: Convert dictionary list to a DataFrame

```python

info_df = pd.DataFrame(ll)
```

# Step 3: Merge with original DataFrame

```python

df_final = pd.concat([df, info_df], axis=1)
```

# Step 4: Optionally convert numeric fields

```python

for col in df_final.columns:
    df_final[col] = pd.to_numeric(df_final[col], errors='ignore')
```
