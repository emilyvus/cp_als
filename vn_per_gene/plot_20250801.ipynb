```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
infile = "20250801/variants_count_per_gene.csv"
df = pd.read_csv(infile)
df.head()
```

```python
df.plot(kind='bar', title='Variant Count Per Gene', y="variant_count", x="gene")
```

```python
infile = "20250801/variants_count_per_genome.csv"
df = pd.read_csv(infile)
df.head()
```

```python
# Create the bar chart
plt.bar(range(df.shape[0]), df['variant_count'])

# Add labels and title
plt.xlabel('Genome ID')
plt.ylabel('Variant Count')
plt.title('Variant Count Per Genome - Across All Genes')

# Display the chart
plt.show()
```

```python
infile = "20250801/variants_avg_per_genome_top_genes.csv"
df = pd.read_csv(infile)
df.head()
```

```python
# Create the bar chart
plt.bar(range(df.shape[0]), df['variant_avg'])

# Add labels and title
plt.xlabel('Genome ID')
plt.ylabel('AVG Variant Count')
plt.title('AVG Variant Count Per Genome - Across Top 10 Genes')

# Display the chart
plt.show()
```
