```python
import pandas as pd
from mylib import vn_genome_ids
import matplotlib.pyplot as plt
```

# The table of all genes and statistics of genome variant counts 

```python
infile = "data/all.csv"
df = pd.read_csv(infile)
df = df[vn_genome_ids + ['gene']]
```

```python
sdf=df.groupby(["gene"]).sum()
```

```python
sdf.head()
```

# Compute statistics of genome variant counts

### Per genome stats

```python
genome_stats= sdf.describe()
```

```python
genome_stats.head()
```

```python
genome_stats_df = pd.concat([sdf,genome_stats],axis=0)
```

```python
genome_stats_df
```

### Per gene stats

```python
gene_df = sdf.transpose().describe()
```

```python
gene_df.transpose()
```

```python
rdf = pd.concat([sdf,gene_df.transpose()],axis=1)
```

# Plot stacked bar chart of total variant counts per gene

```python
infile = "data/all.csv"
df = pd.read_csv(infile)
df = df[['gene','POS']]
```

```python
gdf=df.groupby(["gene"]).count()
```

```python
gdf
```

```python
gdf.reset_index(inplace=True)
```

```python
gdf.rename(columns={"POS":"variant_count"},inplace=True)
gdf=gdf.sort_values(by=["variant_count"],ascending=False).reset_index(drop=True)
```

```python
gdf
```

```python
gdf.plot(kind='bar', title='Variant Count Per Gene', y="variant_count", x="gene")
```

### Top 5 genes in terms of total variant counts across all genomes

```python
top_genes = list(gdf['gene'][0:5])
top_genes
```

```python
infile = "data/all.csv"
df = pd.read_csv(infile)
df = df[vn_genome_ids]
```

## 1: Get the top genomes and bottom genomes in terms of total variant counts across all genes

```python
hdf=pd.DataFrame()
hdf['count'] = df.sum()
```

```python
hdf.reset_index(inplace=True)
```

```python
hdf.head()
```

```python
mean,std = hdf['count'].mean(),hdf['count'].std()
mean,std
```

```python
top_genomes = list(hdf[hdf['count'] > mean+std]['index'])
bottom_genomes = list(hdf[hdf['count'] < mean-std]['index'])
```

```python
len(top_genomes)
```

```python
len(bottom_genomes)
```
