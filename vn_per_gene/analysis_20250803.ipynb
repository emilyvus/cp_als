## I expected a more equal distribution (flat) but this is clearly skewed towards a subset in the population. We can also display the different genes using different colours in this bar chart (i.e make it into a stacked bar chart). If you want to make a stacked bar chart, to avoid having too many samples, take a subset of samples with >1000 variants (subset 1), and a subset of samples with <600 variants (subset 2), and make the charts for these. I used 1000 and 600 arbitrarily, you can use (mean + 2SD) and (mean - 2SD) as the threshold to select your subsets.
