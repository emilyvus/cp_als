# Plot the bar chart of total variant count per gene

```python
infile = "data/all.csv"
df = pd.read_csv(infile)
df = df[['gene', 'POS']]
df.head()
```

```python
gdf=df.groupby(["gene"]).count()
```

```python
gdf.head()
```

```python
gdf.reset_index(inplace=True)
```

```python
gdf.rename(columns={"POS":"Variant Count"},inplace=True)
gdf=gdf.sort_values(by=["Variant Count"],ascending=False).reset_index(drop=True)
```

```python
gdf.head()
```

```python
#gdf.plot(kind='bar', title='Variant Count Per Gene', y="variant_count", x="gene")
ax = gdf.plot(kind='bar', y="Variant Count", x="gene", figsize=(8,6), fontsize=9)
ax.set_title('Total Variant Count Per Gene')
#ax.set_ylabel('Variant Count')
ax.set_xlabel('Gene Name')
plt.xticks(rotation=-60) # Rotate x-axis labels for better readability
#plt.legend(title='Total Variant Count')
#plt.legend(title='Genes', loc="lower left")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

## Stacked Bar Chart

```python
infile = "data/all.csv"
df = pd.read_csv(infile)
df = df[vn_genome_ids]
df.head()
```

## First, get the top genomes and bottom genomes in terms of total variant counts across all genes

```python
hdf = pd.DataFrame()
hdf['count'] = df.sum()
```

```python
hdf.reset_index(inplace=True)
```

```python
hdf.head()
```

```python
mean, std = hdf['count'].mean(), hdf['count'].std()
mean, std
```

```python
top_genomes = list(hdf[hdf['count'] > mean + std]['index'])
bottom_genomes = list(hdf[hdf['count'] < mean - std]['index'])
```

```python
len(top_genomes)
```

```python
len(bottom_genomes)
```

```python
infile = "data/all.csv"
df = pd.read_csv(infile)
df = df[vn_genome_ids + ['gene']]
sdf=df.groupby(["gene"]).sum()
```

```python
sdf.head()
```

```python
tgdf = sdf.loc[top_genes]
```

```python
tgdf.head()
```

```python
tgdf_top_genomes = tgdf[top_genomes]
```

```python
tgdf_bot_genomes = tgdf[bottom_genomes]
```

```python
tgdf_top_genomes
```

```python
tgdf_bot_genomes
```

```python
tgdf_top_genomes.index
```

```python
tgdf_bot_genomes.index
```

```python
ax = tgdf_top_genomes.transpose().plot.bar(stacked=True, figsize=(8, 6))
ax.set_title('Top Genes - Top Genomes: Variant Count')
ax.set_ylabel('Variant Count')
ax.set_xlabel('Genome Name')
plt.xticks(rotation=-45) # Rotate x-axis labels for better readability
plt.legend(title='Genes', loc="lower left")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
ax = tgdf_bot_genomes.transpose().plot.bar(stacked=True, figsize=(8, 6))
ax.set_title('Top Genes - Bottom Genomes: Variant Count')
ax.set_ylabel('Variant Count')
ax.set_xlabel('Genome Name')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xticks(rotation=-45) # Rotate x-axis labels for better readability
plt.legend(title='Genes', loc="lower left")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
