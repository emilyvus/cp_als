```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import random
# Example: generate some synthetic data
np.random.seed(42)
df1 = pd.DataFrame({
    'feature1': np.random.normal(0, 1, 100),
    'feature2': np.random.normal(1, 3, 100),
    'feature3': np.random.normal(2, 5, 100)
})

df2 = pd.DataFrame({
    'feature1': np.random.randint(10, 31, 100),
    'feature2': np.random.randint(15, 32, 100),
    'feature3': np.random.randint(16, 25, 100)
})
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2])
```

```python
df.shape
```

```python
# Step 1: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
# Step 2: Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Step 3: Visualize the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[0:100, 0], X_pca[0:100, 1], c='blue', alpha=0.6, edgecolors='k', label='GBR')
plt.scatter(X_pca[100:200, 0], X_pca[100:200, 1], c='red', alpha=0.6, edgecolors='k', label='KHV')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('PCA - First Two Components')
plt.grid(True)
plt.show()
```
